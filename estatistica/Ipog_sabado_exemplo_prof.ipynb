```python
# importação da biblioteca para separar os dados de forma estratificada
from sklearn.model_selection import train_test_split
```

```python
# separação os dados de forma estratifica pela coluna "Produto" proporcional a 20%
X_train, X_test, y_train, y_test = train_test_split(dados.drop('Produto', axis=1), 
                                                    dados['Produto'], 
                                                    test_size=0.2, 
                                                    random_state=1,
                                                    stratify=dados['Produto'])

# os dados de interesse estão em X_test e y_test.
```

```python
# combinando os dois resultados
estratificado = X_test
estratificado['Produto'] = y_test
estratificado[['Produto', 'tempo-entrega-real']]
```
